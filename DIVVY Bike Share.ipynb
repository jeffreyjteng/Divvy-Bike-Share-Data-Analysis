```python
df.isna().sum()
```

```python
df=df.dropna(subset=['end_station_name','end_station_id'])
```

```python
df=df.dropna(subset=['start_station_name','start_station_id'])
```

```python
df.isna().sum()
```

```python
df['started_at_time'] = df['started_at'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
```

```python
df['ended_at_time'] = df['ended_at'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
```

```python
df['duration_time']=df['ended_at_time']-df['started_at_time']
```

```python
negatives = df['duration_time'] < pd.Timedelta(0)
df = df.loc[~negatives]
```

```python
df['duration_time'] = df['duration_time'].apply(lambda x: str(x)[-8:])
```

```python
df.to_csv("divvy_2021-2022.csv",index=False)
```
